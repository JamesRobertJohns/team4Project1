```python
df = pd.read_csv("../data/output_1.csv")
x = np.array(df["ArraySize"])
y = np.array(df["TotalKeyCmp"])
y2 = np.array(df["KeyCmpInsertionSort"])
fig = plt.figure(figsize=(12,10))

plt.plot(x, y, label='Total Key Comparisons')
plt.plot(x, y2, label='Key Comparisons by InsertionSort()')

plt.xlabel('Size of Array')
plt.ylabel('Number of Key Comparisons')
plt.title('Growth of Key Comparisons against Array Size')
plt.legend()
plt.show()
```

```python
df = pd.read_csv("../data/output_1.csv")
x = np.array(df["ArraySize"])
y = np.array(df["TotalKeyCmp"])
fig = plt.figure(figsize=(12,10))
y2 = 2*20*x + 2*np.log2(x/20)
y1 = 1/2*20*x + 1/2*np.log2(x/20)

plt.plot(x, y, label='Total Key Comparisons')
plt.plot(x, y2, label='c_2 = 2')
plt.plot(x, y1, label="c_1 = 1/2")

plt.xlabel('Size of Array')
plt.ylabel('Number of Key Comparisons')
plt.title('Growth of Key Comparisons against Array Size')
plt.legend()
plt.show()
```
