```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os

sns.set()
```

```python
df = pd.read_csv("../data/output_1.csv")
x = np.array(df["ArraySize"])
y = np.array(df["TotalKeyCmp"])
y2 = np.array(df["KeyCmpInsertionSort"])
fig = plt.figure(figsize=(12,10))

plt.plot(x, y, label='Total Key Comparisons')
plt.plot(x, y2, label='Key Comparisons by InsertionSort()')
plt.plot(x, 21*x, label='O(kn)', linestyle='--', color='red')

plt.xlabel('Size of Array')
plt.ylabel('Number of Key Comparisons')
plt.title('Growth of Key Comparisons against Array Size')
plt.legend()
```

```python
label = ["Array Sized 1 Thousand", "Array Sized 10 Thousand", "Array Sized 100 Thousand", "Array Sized 1 Million", "Array Sized 10 Million"]
fig = plt.figure(figsize=(12,10))

i = 0
fname = "../data/ciii_"
bname = ".csv"
for i in range(5):
    df = pd.read_csv(fname+str(i+1)+bname)
    x = np.array(df["S"])
    y = np.array(df["TotalKeyCmp"])
    plt.plot(x, y, label=label[i])

plt.xlabel('Value of S')
plt.ylabel('Number of Key Comparisons')
plt.title('Growth of Key Comparisons against S for Various Array Sizes')
plt.legend()
plt.show()
```
